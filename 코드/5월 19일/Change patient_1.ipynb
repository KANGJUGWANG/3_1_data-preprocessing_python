```python
data= pd.read_csv("Data/patient_1.csv")# 기존 df
new_data = pd.DataFrame()# 새로 구성할 df
```

```python
# 1. 결측값 포함 행 제거
data = data.dropna()
```

```python
#2. 검사코드, 검사결과값(표시), 진료과 코드, 진료과 코드NM, 성볅코드 NM 제거
data.columns
```

```python
new_data= data.drop(['검사코드','검사결과값(표시)','진료과코드','진료과코드NM','성별코드NM'], axis=1)
```

```python
new_data
```

```python
#4. 입실 일자, 접수일자, 퇴실 일자 포멧 변경
# 숫자 -> 문자열 변환
new_data['입실일자'] = new_data['입실일자'].astype(str)
new_data['접수일자'] = new_data['접수일자'].astype(int).astype(str)
new_data['퇴실일자'] = new_data['퇴실일자'].astype(str)
# datetime형으로 변환
new_data['입실일자'] = pd.to_datetime(new_data['입실일자'])
new_data['접수일자'] = pd.to_datetime(new_data['접수일자'])
new_data['퇴실일자'] = pd.to_datetime(new_data['퇴실일자'])
```

```python
#5. 입실 시각, 접수시각(?), 퇴실 시각 포멧 변경
# 문자열 변경및 6자리고정
new_data['입실시각'] = new_data['입실시각'].astype(int).astype(str).str.ljust(4,fillchar='0')
new_data['접수시각'] = new_data['접수시각'].astype(int).astype(str).str.ljust(4,fillchar='0')
new_data['퇴실시각'] = new_data['퇴실시각'].astype(str).str.ljust(4,fillchar='0')
# datetime형으로 변환
new_data['입실시각'] = pd.to_datetime(new_data['입실시각'],format="%H%M", errors='coerce').dt.time
new_data['접수시각'] = pd.to_datetime(new_data['접수시각'],format="%H%M", errors='coerce').dt.time
new_data['퇴실시각'] = pd.to_datetime(new_data['퇴실시각'],format="%H%M", errors='coerce').dt.time
new_data['입실시각'] 
```

```python
#6 검사나이 x대/y세로 그룹화
def age_group(age):
    if age <=19:
        return '10대'
    elif age <=29:
        return '20대'
    elif age <=39:
        return '30대'
    elif age <=49:
        return '40대'
    elif age <=59:
        return '50대'
    elif age <=69:
        return '60대'
    elif age <=79:
        return '70대'
    elif age <=89:
        return '80세 '
    else :
        return '80세 이상'
```

```python
## 검사나이 변경
new_data['검사나이'] = new_data['검사나이'].apply(age_group)
```

```python
#7. '입원기간'열 추가 
new_data['입원기간'] = new_data['퇴실일자']-new_data['입실일자']
```

```python
#8. 환자 id(생성규칙모름) 추가
# 가설
# 1. 환자 dl에서 인덱스순서가 3,4인 내용은 제거하여 문자열을 생성한다
# 2. 생성된 문자열의 길이가 8미만이라면 부족한 만큼 입실 날자에서 가지고 온다 
# 3. 제거된 인덱스 수에 따라서(2 혹은 1 아니면 0) 입실일자[-2+(제거된 인덱스 수):]의 규칙으로 8자리를 생성한다
def set_id(data):
    ## dl의 최대 길이 10, 최소 길이 4
    dl_s = str(data['연구번호'])
    if len(dl_s)>=5:## 인덱스 2개 제거 가능 
        index =2
    elif len(dl_s)==4:# 인덱스 1개 제거 가능
        index =1
    dl = dl_s[:3] + dl_s[5:]
    
    if len(dl)==8:## 제거된 길이가 8이라면 
        return dl
    else:## 길이가 8보다 작다면 부족한 만큼 입실일자 추가
        return dl+str(data['입실일자']).replace('-','')[2-index : 8-len(dl)+(2-index)]
        
    
new_data['환자ID']=new_data.apply(set_id, axis=1)
    
```
