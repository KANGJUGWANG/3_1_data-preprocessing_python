```python
import pandas as pd
import numpy as np
```

```python
data_1  = pd.read_csv("Data/new_patient_1.csv")
data_2  = pd.read_csv("Data/patient_2.csv", encoding = "euc-kr")
# 공통 
# 입실일자 2021-02-20
# 입실 시각 13:50:00
# 검사 나이 80세
# 퇴실일자 2021-03-02
# 환자 생년월일 193909
# 성별 M
# 입원 기간 10 days
```

```python
print(data_1.columns,"\n",data_2.columns )
```

```python
set_1=data_1[data_1['환자생년월']==193909]
set_2=data_2[data_2['환자생년월']==193909]
print(set_1, set_2)
```

```python
subset_data = pd.concat([set_1, set_2], axis=0, ignore_index=True)
subset_data.drop(['접수일자','접수시각'], axis=1)
```

```python
subset_data.to_csv("subset_data.csv",index = False,encoding = 'euc-kr')
```
