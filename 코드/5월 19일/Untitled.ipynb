```python
df = pd.DataFrame({'문항 1' : ['1','2,,','3,','3'], '문항 2' : [',,2',',2','3',',2']})
print(df)
df['문항 1'] = df['문항 1'].str.rstrip(",")
df['문항 2'] = df['문항 2'].str.lstrip(",")
print(df)
```

```python
df = pd.read_csv("C:/Users/pc502/Downloads/health.csv")
df
```

```python
diet_columns = ['다이어트경험_답변1', '다이어트경험_답변2', '다이어트경험_답변3', '다이어트경험_답변4']
df[diet_columns] = df[diet_columns].replace(r'^\s*$', np.nan,regex=True)
df
```

```python
## 단답형과 같이 행에서 1개만 값을 가질때 낮은열을 우선순위로 추출(이후 값 무시)
df['묶기'] = df[diet_columns].bfill(axis=1).iloc[:,0]
df
```

```python
## 행별 독립변수(NaN 제외) 그룹화
df['하나의 항목'] = df[diet_columns].stack().groupby(level = 0).agg(list)
df
```
