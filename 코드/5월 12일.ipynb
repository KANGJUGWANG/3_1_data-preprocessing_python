```python
# 데이터 프레임 조건 적용
data = {'Numeric1': [1, np.nan, 3,4],
        'Numeric2': [5, 6, np.nan,8.2],
        'Category': ['doge',np.nan, 'cat','dog']
       }
df = pd.DataFrame(data)
print(df[(df['Numeric2']>=6) &(df['Numeric2']<=10)])
print(df[(df['Numeric2']>=6) |(df['Numeric2']<=10)])
```

```python
data = {
    "나이" : [22,35,16,55,42]
}
df = pd.DataFrame(data)
```

```python
## 사용자 임의 함수 생성
def categorize(data):
    if 0< data<=10:
        return '10대'
    elif data<=20:
        return '20대'
    elif data<=30:
        return '30대'
    else:
        return '30대 이상'
```

```python
df['검사 나이'] = df['나이'].apply(categorize)
print(df)
```

```python
## 특정 문자열(문자)를 포함한 행 찾기 
data = {'Numeric1': [1, np.nan, 3,4],
        'Numeric2': [5, 6, np.nan,8.2],
        'Category': ['dog',np.nan, 'cat','dogg']
       }
df=pd.DataFrame(data)
filter_df = df[df['Category']=='dog']
filter_df_contains = df[df['Category'].str.contains('d',na=False)]
print(filter_df,'\n',filter_df_contains)
```

```python
df= pd.DataFrame({'variable':['value1-value2-value3',"value3-value4-value5","vvdfvalue165656"]})
# 특정 문자가 포함된 문장전체를 뒤로 이동하면서 찾음 
df['new_variable'] = df['variable'].replace(r'value1.*','value1', regex=True)
print(df)
# 특정 문자열을 포함하면 문자열 전체를 특정 문자열로 변경
df['new_variable2'] = df['variable'].str.extract(r'(value1)')
print(df)
df['new_variable3'] = df['variable'].str.extract(r'(value2)')
print(df)
```

```python
#문자열 분할하기 (토큰 작업)
df= pd.DataFrame({'variable':['value1-value2-value3',"value4-value5","vvdfvalue165656"]})
split_df = df['variable'].str.split('-',expand=True)
split_df.columns=['value1','value2','value3']
split_df
```
